```python
import pandas as pd
```

```python
"""
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. 
Определите какому фильму было выставлено больше всего оценок 5.0.
"""
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')

movie_rates = ratings.merge(movies, on='movieId', how='left')
(movie_rates.groupby(['title','rating']).count()
 .sort_values(['rating','userId'], ascending=[False,False])
 .reset_index()[['title','rating','userId']].head(1))
 
```

```python
"""По данным файла power.csv посчитайте суммарное потребление стран Прибалтики 
(Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. 
Не учитывайте в расчетах отрицательные значения quantity."""
data = pd.read_csv('power.csv')

data.query("(category in ['4','12','21'])\
           and 2005<=year<=2010 and quantity>0\
           and country in ['Lithuania', 'Latvia', 'Estonia']").sum()['quantity']
```
