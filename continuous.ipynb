Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.



```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4,4,200)
plt.plot(x, sts.norm.pdf(x, loc=0, scale=0.7))
plt.plot(x, sts.norm.pdf(x, loc=0, scale=1.5))
plt.plot(x, sts.norm.pdf(x, loc=1, scale=2))
plt.plot(x, sts.norm.pdf(x, loc=0, scale=0.2))
plt.plot(x, sts.norm.pdf(x, loc=-2, scale=0.1))
plt.show()
```

```python
x = np.linspace(0,4,200)
plt.plot(x, sts.expon.pdf(x, loc=0, scale=0.3))
plt.plot(x, sts.expon.pdf(x, loc=0, scale=1.5))
plt.plot(x, sts.expon.pdf(x, loc=1, scale=2))
plt.plot(x, sts.expon.pdf(x, loc=0, scale=0.2))
plt.plot(x, sts.expon.pdf(x, loc=-2, scale=0.1))
plt.show()
```

```python
x = np.linspace(-4,4,200)
df = 2.74
plt.plot(x, sts.t.pdf(x, df, loc=0, scale=0.3))
plt.plot(x, sts.t.pdf(x, df, loc=0, scale=1.5))
plt.plot(x, sts.t.pdf(x, df, loc=1, scale=2))
plt.plot(x, sts.t.pdf(x, df, loc=0, scale=0.2))
plt.plot(x, sts.t.pdf(x, df, loc=-2, scale=0.1))
plt.show()
```

Непрерывная случайная величина задана плотностью распределения:$$ f(\xi) = C, \xi \in [0;5];f(\xi) = 0, \xi \notin [0;5]$$Найдите C, математическое ожидание $\xi$ а также вероятность попадания в отрезок $[3.5;7]$, т.е. P(3.5 < $\xi$ < 7).



```python
#Плотность распределения равномерной непрерывной случайной величины равна 1/(b-a) на заданом промежутке а-b
С = 1/5
С
```

```python
#Математическое ожидание M = (a+b)/2
M = 5/2
M
```

```python
# Вероятность попадания в отрезок от 5 до 7 равна нулю. Вероятность попадания в отрезок 3.5-5:
P = (5-3.5)/5
P
```
