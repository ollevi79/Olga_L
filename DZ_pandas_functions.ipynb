```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
```

```python
def classification(rating):
    if rating <= 2:
        return 'низкий рейтинг'
    if 2 < rating <=4:
         return 'средний рейтинг'
    if rating >= 4.5:
         return 'высокий рейтинг'
```

```python
ratings['class'] = ratings['rating'].apply(classification)
pd.concat([ratings.loc[ratings.rating == 1].head(1),ratings.loc[ratings.rating == 2].head(1),ratings.loc[ratings.rating == 3].head(1),ratings.loc[ratings.rating == 4].head(1),ratings.loc[ratings.rating == 5].head(1)])
```

```python
keywords = pd.read_csv('keywords.csv')
```

```python
def find_city(keyword):
    import re
    geo_data = {'Центр': ['Москва', 'Тула', 'Ярославль'],
              'Северо-Запад': ['Петербург', 'Псков', 'Мурманск'],
              'Дальний Восток': ['Владивосток', 'Сахалин', 'Хабаровск']}
    for key, value in geo_data.items():
        for city in value:
            l = len(city)
            pattern = r'\b'+city[0:l-1]+r'\w*'
            if re.findall(pattern,keyword,flags=re.IGNORECASE) != []:
                    return key
    return 'Undefined'
        
keywords['region'] = keywords['keyword'].apply(find_city)
pd.concat([keywords.loc[keywords.region == 'Северо-Запад'].head(),keywords.loc[keywords.region == 'Центр'].head(),keywords.loc[keywords.region == 'Дальний Восток'].head(),keywords.loc[keywords.region == 'Undefined'].head()])
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
```

```python
def production_year(title):
    import re
    years = list(range(1950,2011))
    for value in years:
            pattern = r'('+str(value)+r')'
            if re.findall(pattern,title,flags=re.IGNORECASE) != []:
                    return value
    return '1900'
        
movies['year'] = movies['title'].apply(production_year)
movie_rates = ratings.merge(movies, on='movieId', how='left')
movie_rates.groupby(['year']).mean('rating').sort_values(['rating'], ascending=[False]).reset_index()[['year','rating']]
```
