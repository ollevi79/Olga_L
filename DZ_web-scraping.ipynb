```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
KEYWORDS = ['powerpoint','интернет']
```

```python
req = requests.get('https://habr.com/ru/all/')
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
posts = soup.find_all('article', class_='post')
```

```python
df = pd.DataFrame()
for post in posts:
    title = post.find('a', class_='post__title_link')
    date = post.find('span', class_='post__time')
    text = post.find('div', class_='post__text')
    string = str(title).lower()+str(text).lower()
    ref = title.attrs.get('href')
    for k in KEYWORDS: 
        search = string.find(k)
        if search != -1: 
            row = {'дата': date, 'заголовок': title, 'ссылка': ref}
            df = pd.concat([df, pd.DataFrame([row])])
df
```
