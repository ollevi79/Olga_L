Посчитайте значение первой и второй производной функции:    

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.    

Уровень 1:     
Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
from scipy.misc import derivative
from sympy import *
import numpy as np
```

```python
x = symbols('x')

def f(x):
    return x**5 + 4*np.sin(2*x) + np.cos(3*x+3)

print(derivative(f, 1.0, dx=1e-6, n=1))
print(derivative(f, 1.0, dx=1e-6, n=2))
```

```python
import matplotlib.pyplot as plt
```

```python
x = symbols('x')
x = np.linspace(-5, 5, endpoint=True)

def f(x):
    return np.sin(2*x+1)**5   

y1 = f(x)         
y2 = derivative(f, x, dx=1e-6, n=1)
y3 = derivative(f, x, dx=1e-6, n=2)

fig, ax = plt.subplots()                        
ax.plot(x, y1, color="blue", label="y(x)")      
ax.plot(x, y2, color="red", label="y'(x)")      
ax.plot(x, y3, color="green", label="y''(x)")   
ax.set_xlabel("x")                              
ax.set_ylabel("y")                              
ax.legend()                                     

plt.show()                                      
```
